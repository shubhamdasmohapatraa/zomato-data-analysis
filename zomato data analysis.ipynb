```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#create a dataframe 
dataframe=pd.read_csv('Zomato data .csv')
print(dataframe)
```

```python
dataframe 
```

```python
#convert the datatype of  coloumn rate that is i have to 4.1/5 to only 4.1 
def handelRate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handelRate)
print(dataframe.head())

    
```

```python
dataframe.info()
```

```python
#type of returant 
dataframe.head(10)
```

```python
sns.countplot(data=dataframe,x='listed_in(type)')
plt.xlabel('type of resturant')
plt.ylabel('count')
plt.title('Distribution of resturant types')
plt.show()
```

so conclusion is majority of the customers are from dinning catagory

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()

# Convert the Series to a DataFrame for plotting
result = pd.DataFrame({'votes': grouped_data})

# Plot the data
plt.plot(result.index, result['votes'], color="green", marker="o")

# Add labels and title
plt.xlabel("Type of Restaurant", color="green", fontsize=20)
plt.ylabel("Votes", color="green", fontsize=20)
plt.title("Total Votes by Type of Restaurant")

# Show the plot
plt.show()
```

so conclusion has recived max votes

```python
# Define the number of bins (e.g., 10 bins)
num_bins = 10

# Plot the histogram
plt.hist(dataframe['rate'], bins=num_bins, edgecolor='black')

# Add title and labels
plt.title("Rating Distribution")
plt.xlabel("Rate")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x ="online_order",y="rate",data =dataframe)
```

```python
# Create the pivot table
pivot_table = dataframe.pivot_table(
    index='listed_in(type)', 
    columns='online_order',
    aggfunc='size',
    fill_value=0
)

# Create the heatmap
sns.heatmap(
    pivot_table, 
    annot=True, 
    cmap="YlGnBu",  # Correct color map parameter
    fmt='d'
)

# Add labels and title
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.title("Heatmap of Restaurant Types and Online Orders")

# Show the plot
plt.show()
```
